#Q1. Q1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?

CSS Flexbox vs. CSS Grid:
1. Layout Structure
Flexbox is a one-dimensional layout system, meaning it works in either a row or a column at a time. It is designed for arranging items along a single axis (horizontal or vertical).

Grid, on the other hand, is a two-dimensional layout system, allowing you to control both rows and columns simultaneously. This makes it ideal for more complex layouts.

2. Alignment & Positioning
Flexbox provides powerful alignment capabilities along a single axis. It makes it easy to distribute space, align elements, and handle flexible sizing within a row or column.

Grid allows for precise placement of items using grid lines and areas, giving you more control over both dimensions.

3. Complexity & Usage
Flexbox is simpler and best suited for arranging smaller components, like a navigation bar, buttons, or form elements.

Grid is more complex but extremely powerful for structuring an entire webpage layout or designing grid-based sections like image galleries, dashboards, and cards.

4. Responsiveness & Adaptability
Flexbox is great for dynamic content that adjusts based on the available space, making it useful for fluid, flexible designs.

Grid is better when you need a rigid structure with defined tracks, though it can also be responsive with features like auto-fit and auto-fill.

#Q2. Q2. Explain the role of the following key properties in the Flexbox layout mode justify-conten align-item/ga) flex-directio flex-wrap.

1. justify-content
This property controls horizontal alignment (main axis) of flex items within the container.

Values:

flex-start (default) → Items align at the start of the container.

flex-end → Items align at the end of the container.

center → Items are centered along the main axis.

space-between → Items spread out, with space only between them.

space-around → Equal space around each item.

space-evenly → Equal space between and around items.

2. align-items
This property controls vertical alignment (cross axis) of flex items within the container.

Values:

stretch (default) → Items stretch to fill the container height.

flex-start → Items align at the top.

flex-end → Items align at the bottom.

center → Items align in the center.

baseline → Items align based on their text baselines.

3. gap
This property defines the spacing between flex items.

Unlike margin, gap applies only between items, making layout management cleaner.

Example:

css
Copy
Edit
display: flex;
gap: 20px;
This creates a 20px space between all flex items.

4. flex-direction
This property determines the main axis direction (row or column).

Values:

row (default) → Items are placed in a horizontal row (left to right).

row-reverse → Items are placed in a horizontal row but reversed (right to left).

column → Items are placed in a vertical column (top to bottom).

column-reverse → Items are placed in a vertical column but reversed (bottom to top).

5. flex-wrap
This property controls whether flex items wrap to the next line when there isn’t enough space.

Values:

nowrap (default) → Items stay in a single line and shrink if needed.

wrap → Items move to the next line when space runs out.

wrap-reverse → Items wrap to the next line, but in reverse order.
